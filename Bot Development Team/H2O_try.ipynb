```python
from pyngrok import ngrok

def start_ngrok_tunnel(port):
    try:
        ngrok_tunnel = ngrok.connect(port)
        ngrok_url = ngrok_tunnel.public_url
        ngrok.set_auth_token("Your Auth token here")
        print(f"Ngrok tunnel is active. Public URL: {ngrok_url}")
        return ngrok_url

    except Exception as e:
        print(f"Error: Ngrok tunnel couldn't be established. {e}")
        return None

if __name__ == "__main__":
    # Replace 10101 with the port number of your localhost server.
    local_server_port = 10101
    public_url = start_ngrok_tunnel(local_server_port)
    !make -f Makefile llmstudio
```
